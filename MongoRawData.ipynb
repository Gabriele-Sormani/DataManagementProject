```python

import json
```

```python
from pymongo import MongoClient
client = MongoClient('localhost', 27017)
db = client.RawDataDb
```

```python
with open("AqiData.json") as f:
    data = json.loads(f.read())

aqi = db.AqiData
aqi.insert_many(data)
```

```python
stations_code = [98497, 107293, 98494, 98485, 460336, 69364, 414133, 21373, 156526, 67897, 112273, 460762,
                 116353, 246697, 196504, 74863, 193501]
weather = db.WeatherData

for code in stations_code:
    filename = 'MeteoData'+ str(code)
    
    with open(filename+".json") as f:
        data = json.loads(f.read())
        for obj in data:
            weather.insert_one(data[obj])
    
    
    
```

```python

traffic = db.TrafficData

for i in range(1, 11):
    filename = 'TrafficData'+str(i)
    with open(filename+".json") as f:
        data = json.loads(f.read())
        
        null_count = 0
        for element in data:
            sourceUpdt = element['sourceUpdated']
            for e in element['results']:
                e['datetime'] = sourceUpdt
                loc = e['location']
                
                if 'description' not in loc.keys():
                    null_count += 1
                else:
                    if len(loc['description']) > 100:
                        e['location']['description'] = loc['description'][0:100]
                traffic.insert_one(e)
        
```
